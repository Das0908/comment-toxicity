```python
x = df['comment_text']
y = df[df.columns[2:]].values # .values changes it into a numpy array
```

```python
MAX_FEATURES = 200000 #number of words in the vocab  , larger number - larger model
```

```python
vectorizer = TextVectorization ( max_tokens = MAX_FEATURES ,
                                output_sequence_length = 1800, # input length of model
                                output_mode = 'int')
```

```python
vectorizer.adapt(x.values)
```

```python
vocab = vectorizer.get_vocabulary()  # integer values assigned to the words are their indices
```

```python
vectorizer('hello , life is shit')[:4]
```

```python
vectorized_text = vectorizer(x.values)
```

```python
import json
with open('vocab.json', 'w') as f:
    json.dump(vocab, f)
```

```python
vectorized_text
```

```python
# tf data pipeline  MCSBP - map , cache , shuffle , batch , prefetch
dataset = tf.data.Dataset.from_tensor_slices((vectorized_text , y))
dataset = dataset.cache()
dataset = dataset.shuffle(160000)
dataset = dataset.batch(16)
dataset = dataset.prefetch(8) # prevents bottlenecks
```

```python
train = dataset.take(int(len(dataset)*.7))
val = dataset.skip(int(len(dataset)*.7)).take(int(len(dataset)*.2))
test = dataset.skip(int(len(dataset)*.9)).take(int(len(dataset)*.1))
```

# dl model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout , Bidirectional , LSTM , Embedding , Dense
```

```python
model = Sequential()
# Create the embedding layer
model.add(Embedding(MAX_FEATURES+1, 32))
# Bidirectional LSTM Layer
model.add(Bidirectional(LSTM(32, activation='tanh')))
# Feature extractor Fully connected layers
model.add(Dense(128, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
# Final layer
model.add(Dense(6, activation='sigmoid'))
model.compile(loss='BinaryCrossentropy', optimizer='Adam')
```

```python
model.summary()
```

```python
history = model.fit(train, epochs=1, validation_data=val)
```

model evaluation

```python
from tensorflow.keras.metrics import Precision , Recall , CategoricalAccuracy
```

```python
pre = Precision()  #( these 3 , update their values over time , good when we have batch data )
re = Recall()
acc = CategoricalAccuracy()
```

```python
for batch in test.as_numpy_iterator():
  #unpack the batch
  X_true , Y_true = batch
  #make predictions
  Y_hat = model.predict(X_true)

  #flatten the predictions ( convert to a 1d array)
  Y_true = Y_true.flatten()
  Y_hat = Y_hat.flatten()

  pre.update_state(Y_true , Y_hat)
  re.update_state(Y_true , Y_hat)
  acc.update_state(Y_true , Y_hat)
```

```python
print(f'precision : {pre.result().numpy()} , recall : {re.result().numpy()} , Accuracy : {acc.result().numpy()}')
```

```python
def score_comment(comment):
  vectorized_comment = vectorizer([comment])
  result = model.predict(vectorized_comment)

  text = ''
  for idx , col in enumerate(df.columns[2:]):
    text += '{}: {}\n'.format(col , result[0][idx] > 0.5)

  return text
```

```python
interface = gr.Interface(
    fn=score_comment,
    inputs=gr.Textbox(lines=2, placeholder='Comment to score'),
    outputs=gr.Text()
)
```
