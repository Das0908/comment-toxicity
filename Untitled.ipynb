# preprocessing (using tokenization )

```python
from tensorflow.keras.layers import TextVectorization
```

```python
help(tf.keras.layers.TextVectorization) #  converts a word to a unique number (tokenization)
```

```python
df.columns
```

```python
df[df.columns[2:]]
```

```python
x = df['comment_text']
y = df[df.columns[2:]].values # .values changes it into a numpy array
```

```python
MAX_FEATURES = 200000 #number of words in the vocab  , larger number - larger model 
```

```python
vectorizer = TextVectorization ( max_tokens = MAX_FEATURES ,
                                output_sequence_length = 1800, # input length of model 
                                output_mode = 'int')
```

```python
vectorizer.adapt(x.values)
```

```python
vectorizer.get_vocabulary()  # integer values assigned to the words are their indices 
```

```python
vectorizer('hello , life is shit')[:4]
```

```python
vectorized_text = vectorizer(x.values)
```

```python
vectorized_text
```

```python
# tf data pipeline  MCSBP - map , cache , shuffle , batch , prefetch 
dataset = tf.data.Dataset.from_tensor_slices((vectorized_text , y))
dataset = dataset.cache()
dataset = dataset.shuffle(160000)
dataset = dataset.batch(16)
dataset = dataset.prefetch(8) # prevents bottlenecks 
```

```python
train = dataset.take(int(len(dataset)*.7))
val = dataset.skip(int(len(dataset)*.7)).take(int(len(dataset)*.2))
test = dataset.skip(int(len(dataset)*.9)).take(int(len(dataset)*.1))
```

# dl model 

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout , Bidirectional , LSTM , Embedding , Dense
```

```python
model = Sequential()
# Create the embedding layer 
model.add(Embedding(MAX_FEATURES+1, 32))
# Bidirectional LSTM Layer
model.add(Bidirectional(LSTM(32, activation='tanh')))
# Feature extractor Fully connected layers
model.add(Dense(128, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
# Final layer 
model.add(Dense(6, activation='sigmoid'))
model.compile(loss='BinaryCrossentropy', optimizer='Adam')
```

```python
model.summary()
```

```python
history = model.fit(train, epochs=1, validation_data=val)
```
